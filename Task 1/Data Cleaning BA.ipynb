```python
import pandas as pd
```

```python
data = pd.read_csv("D:\British_Airways_Reviews.csv")
data.drop(columns=["Unnamed: 0"],inplace=True)
data
```

```python
## Separating Is the trip is verified (i.e) Does the customer actually travelled
data["Verified"] =  data["Reviews"].str.contains("Trip Verified")
data["Verified"] = data["Verified"].replace({True: "Verified", False: "Not Verified"})
data
```

```python
## Removing Trip verified from Reviews
data["Reviews"] = data["Reviews"].str.strip("✅ Trip Verified |")
data
```

```python
## Replacing no by No
data["Recommended Type"] = data["Recommended Type"].replace({"no":"No"})
data
```

```python
## Converting the date into datetime format
data["Date Flown"] = pd.to_datetime(data["Date Flown"])
data
```

```python
## Checking Null values 
data.isnull().value_counts()
```

```python
data.dropna(inplace=True)
data.isnull().value_counts()
```

```python
## Cleaning reviews removing extra characters than a-z then lowering the characters then removing the stopwords and lemmatizing the word
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import re
lemmatizer = WordNetLemmatizer()
reviews = []
for item in data["Reviews"]:
    item = re.sub('[^a-zA-Z]',' ', item)
    item = item.lower()
    tokens = word_tokenize(item)
    item = [lemmatizer.lemmatize(word)for word in tokens if word not in set(stopwords.words("english"))]
    item  = " ".join(item)
    reviews.append(item)
data["Cleaned Reviews"] = reviews
data
```

```python
data.to_csv("D:"+ "/Cleaned_British_Airways_Reviews.csv")
```
